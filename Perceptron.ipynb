Pour cet exemple, je crée un nuage de points répartis aléatoirement dans [-1,1]x[-1,1], ensuite une droite D_cible (en rouge) définie aléatoirement va séparer mon nuage de point en deux sous espaces du plan, d'un côté de la droite j'attribue aux points des labels 1 alors dans l'autre partie j'assigne aux points le label -1.
Une fois mon échantillon de points aléatoires et labelisés créé, je cherche par l'algorithme du perceptron, une droite qui va séparer correctement mes deux nuages de points. Cette droite vise donc à simuler la fonction cible définie par ma droite initiale D_cible (e rouge).
Les rayons en vert représentent les différents essais de mon perceptron au fûr et à mesure des itérations.
La droite bleue est un résultat qui classe correctement tous les points de l'échantillon, l'algorithme s'arrête alors. Notez qu'il y a une infinité de droites possibles et que plus il y a de points dans l'échantillon (variable N) plus le choix des possibles se réduit.
Enfin si mon perceptron échoue au bout d'un nombre max d'itérations (IT), alors il affiche la droite (en jaune) correspondant à la meilleure des solutions rencontrées. Cette droite correspond au minimum des coûts (Taux de points mal labélisés).

La deuxième figure affiche la valeur du coût à chaque itération. Si le perceptron converge, le coût diminuera vers la valeur nulle. Si la convergence n'a pas lieu (phénomène de bagotement) on voit apparaitre le bagotement entre deux valeurs.

Plus il y a de points dans l'echantillon et plus la convergence est tardive en nombre d'itérations mais plus le coût final est faible (en général).

```python
# -*- coding: utf-8 -*-
"""
Created on Sun Mar 08 21:47:56 2015

@author: kim
"""
import numpy as np
import matplotlib.pyplot as plt
from numpy import random
from random import randint



def f(Xa,Xb,Ya,Yb,x):
    return ((((Yb-Ya)/(Xb-Xa))*x + ((Ya*Xb)-(Yb*Xa)/(Xb-Xa))))
    


def f2(Xa,Xb,Ya,Yb,M): #fonction qui sépare les M points de part et d'autre d'une droite et assigne des labels
    cols = len(M[0])
    for i in range(0,cols):
        if (((Yb-Ya)/(Xb-Xa))*M[0,i] + ((Ya*Xb)-(Yb*Xa)/(Xb-Xa))) >= M[1,i]:
            M[2,i] = -1  #Les bels sont dans la troieme ligne de la matrice M
        else:
             M[2,i] = 1
    return M
    
       

def perceptron(W,Xm):
    w = np.asmatrix(W.transpose())
    xm = np.asmatrix(Xm)
    #print("this is w")
    #print(w)
    if w.all() == 0:        
        GXn = np.ones((1,len(Xm[0])))
    else:
        GXn = np.sign(w*xm)
    return np.array(GXn)
    
def Calcul_Cost(P,Xn,N):
    Cost= 0    
    for j in range(len(Xn[0])):
        #print ("deltas P et Xn" , P[0,j]*Xn[3,j] )
        if np.sign(P[0,j]*Xn[3,j]) == -1: 
            Cost = Cost + 1
            #print("Cost" , Cost)
    Cost = float(Cost)/N
    return Cost


def choix_point_KO(Xn,P):
    SubX=np.zeros((3,0))    
    for k in range(len(Xn[0])):
        if P[0,k]<>Xn[3,k]:
            SubX = np.column_stack([SubX, Xn[:3,k]])
    #print("SubX = ", SubX)
    #SubX = SubX[:,1:]
    #print("SubX = ", SubX)
    r = randint(1,len(SubX[0]))-1
    #print(len(SubX[0]), r)
    Xi = SubX[:,r]
    return Xi
    
def play(IT,N):
    #Tirage uniforme dans [-1,1] de N points
    x1 = np.random.uniform(-1,1,N)
    x2 = np.random.uniform(-1,1,N)
    X = np.vstack([x1, x2])
    #on contruit la matrice M des N points avec premiere coordonnée xo egale à 1
    M = np.vstack([X , np.zeros((1,len(X[0])))])
    cols = len(M[0])
    Cost = 0
    PerfIT = 0
    Tab_IT= np.arange(0,IT)

    fig = plt.figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
    ax = fig.add_subplot(1,1,1)
    
    x1, x2 = -2, 2
    y1, y2 = -2, 2
    plt.xlim([x1, x2])
    plt.xlim([y1, y2])
    plt.axis([x1, x2, y1, y2])
    plt.xlabel('X1')
    plt.ylabel('X2')
    plt.title('Perceptron')

    A = [X[0,0],X[0,1],-1.02*X[0,0],-1.02*X[0,1],1.02*X[0,0],1.02*X[0,1]]
    B = [f(X[0,0],X[0,1],X[1,0],X[1,1],X[0,0]), f(X[0,0],X[0,1],X[1,0],X[1,1],X[0,1]),f(X[0,0],X[0,1],X[1,0],X[1,1],-1.02*X[0,0]), f(X[0,0],X[0,1],X[1,0],X[1,1],-1.02*X[0,1]),f(X[0,0],X[0,1],X[1,0],X[1,1],1.02*X[0,0]), f(X[0,0],X[0,1],X[1,0],X[1,1],1.02*X[0,1])]
    plt.plot(A,B,'.r-', linewidth=1)

    M1 = np.zeros((2,1))
    M2 = np.zeros((2,1))
    f2(X[0,0],X[0,1],X[1,0],X[1,1],M)
    for i in range(0,cols): 
        if  M[2,i] == 1:
            M1 = np.column_stack([M1, M[0:2,i]])
            #print("if", M[0:3,i])
        else:
            #print("else", M[0:3,i])
            M2 = np.column_stack([M2, M[0:2,i]])
    #print(M)
    ax.scatter(M1[0,1:], M1[1,1:], marker = 'x', s = 30)
    ax.scatter(M2[0,1:], M2[1,1:], marker = 'o', s = 30)

    #Calcul  des parametres du perceptron

    W = np.zeros((3,1)) #vesteur poids du perceptron
    #Je construis Xn qui est le tableau de mes points avec Xo = 1, Attention la quatrieme ligne de Xn contient le label de chaque point
    Xn = M
    Xn = np.vstack([np.ones(len(Xn[0])),Xn])
    Wtot = np.zeros((4,IT))
 
    for i in range(IT):
        P = perceptron(W,Xn[:3,:]) #tableau d'une ligne des résultats de classification rendus par le perceptron
        #avant de sortir w on le sauvegarde dans un tableau pour retenir au final le parametre qui assure le meilleur taux de bons résultats
        #print("P=",P[:,:20])
        #print("Yn=",Xn[:,:20])
        for p in range(3):
            Wtot[p,i] = W[p]
        Wtot[3,i] =  Calcul_Cost(P,Xn,N)
        #print("Wtot =",Wtot)
        #Result_Cost = np.column_stack([Result_Cost, Calcul_Cost(P,Xn,N)])

        nogood = 0
        for j in range(len(Xn[0])):
            #print (j,"error",P[0,j]*Xn[3,j])
            if (P[0,j]*Xn[3,j]) == -1. :
                nogood = 1
                A = [-1,1]
                B = [-(W[0]/W[2])+(W[1]/W[2]),-(W[0]/W[2])-(W[1]/W[2])]
                plt.plot(A,B,'g-', linewidth= 0.1)
                Ptx=choix_point_KO(Xn,P)
                Yn = Xn[3,j]
                #print("Yn", Yn)
                #print("Ptx" , Ptx)
                #print("Ptx*Yn" , Ptx*Yn) 
                for u in range(3):
                    W[u] =  W[u] + Ptx[u]*Yn
                    #print ("W",W)
                break
        if nogood == 0:
            A = [-1,1]
            B = [-(W[0]/W[2])+(W[1]/W[2]),-(W[0]/W[2])-(W[1]/W[2])]
            plt.plot(A,B,'.b-',linewidth= 1)
            #print("iterations", iterations)
            break
    
        #print("Wtot", Wtot)
    #print("iterations",iterations)
    #print(" calcul du perceptron num", i)
    if nogood == 1:
            index_min=Wtot[3,:].argmin()
            Ws = Wtot[:3,index_min]
            A = [-1,1]
            B = [-(Ws[0]/Ws[2])+(Ws[1]/Ws[2]),-(Ws[0]/Ws[2])-(Ws[1]/Ws[2])]
            plt.plot(A,B,'y-',linewidth= 3)
    PerfIT = i
    #print("tableau des couts", Wtot[3,:])
    best_cost = np.min(Wtot[3,:])
    print("PerfIT = ", PerfIT, "Best_cost1 = ", best_cost )
    #print( Cost)
    #plt.show()
    #******courbe d'evolution du Cost au fur et a mesure des itérations
    fig2 = plt.figure(num=None, figsize=(8, 4), dpi=80, facecolor='w', edgecolor='b')
    a2x = fig2.add_subplot(1,1,1)
    plt.xlabel('IT')
    plt.ylabel('Cost')
    plt.title('Evolution du cout (Cost) en fonction des iterations (IT)')
    plt.plot(Tab_IT,Wtot[3,:],'k-',linewidth= 0.2)
    return PerfIT,best_cost
    
    
    
#******************MAIN***********************************    
N = 100 # nombre de points de l'échantillon
IT = 500 #nombre d'iterations max du perceptron
PlayNbr = 1 #Nombre de d'itération de l'exercice, definition des points et calcul du perceptron
PerfitTOT = 0
PerfITMOY = 0
CostTOT = 0
CostMOY = 0

for k in range(PlayNbr):
    #print("+++++++++++++++++++++++ceci est une nouvelle partie++++++++++++++++++++")
    PerfIT, Cost = play(IT,N)
    PerfitTOT = PerfitTOT + PerfIT
    CostTOT = CostTOT + Cost

PerfITMOY = PerfitTOT/PlayNbr
CostMOY = CostTOT/PlayNbr

print("Nombre d'iterations = ", PerfITMOY)
print( "Cout moyen" , CostMOY )
plt.show()
```

Le perceptron peut ainsi être évalué sur plusieurs jeux de données, afin d'établir une moyenne des itérations avant convergence et une moyenne du coût (Cost).
Biensûr l'execution est un peu plus longue en fonction du paramêtrage, ici 1000 essais (variable PlayNbr).
